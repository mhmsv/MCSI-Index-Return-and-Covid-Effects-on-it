```python
#defining one function to convert quarterly data into monthly,take dataset as input and return reindexed and monthly data
#convert quarterly data to monnthly
#we use this funtion to resample data to monthly, if the data is monthly it just reindex it to its month and do nothing with the data values
def quarterlyToMonthly(data):
    data['Date'] = pd.to_datetime(data['Date']).dt.to_period('M')
    data = data.set_index('Date').resample('M').interpolate()
    return data

```
