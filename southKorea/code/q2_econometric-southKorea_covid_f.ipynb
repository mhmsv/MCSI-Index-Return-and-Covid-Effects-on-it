```python
#crreating regression table
regression_df =  pd.DataFrame({'log_mcsi' : LogDiff(sk_df["MCSI"]),
                               'log_gdp' : LogDiff(sk_df["GDP"]),
                               'log_industrialProduction' : LogDiff(sk_df["Industrial Production"]),
                               'log_cpi' : LogDiff(sk_df["CPI"]),
                               'log_imp_cons_goods' : LogDiff(sk_df["Import of Consumer Goods"]),
                               'log_int_trade_bal' : LogDiff(sk_df["International Trade Balance"]),
                               'log_landPrice' : LogDiff(sk_df["LandPrice"]),
                               'unemprate' : (sk_df["Unemployment"]),
                               'usdkrw':(sk_df["USDKRW"]),
                                'log_consumerSpending':(sk_df["consumer spending"]),
                                'log_bankrupty':(sk_df["bankrupty"]),
                                'interestRate':(sk_df["interestRate"]),
                                'log_toursit_arrival':LogDiff(sk_df["touristArrival"])
                               
                              })
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(regression_df) 
```

```python
#droping land price because of too many NA and also remove data out of july 2002 to june 2022
regression_df = regression_df.drop(['log_landPrice'], axis=1)

```

```python
#applying  regression on the whole period without covid variables:
import statsmodels.formula.api as smf


formula = 'log_mcsi ~ log_gdp + log_industrialProduction + log_cpi + log_imp_cons_goods + log_int_trade_bal + unemprate + usdkrw + log_consumerSpending + log_bankrupty + interestRate + log_toursit_arrival'
est_model2 = smf.ols(formula, regression_df).fit()
est_model2.summary()
```
